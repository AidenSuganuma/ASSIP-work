```python
import cv2
import skimage
import skimage.io as io
import skimage.filters as filters
import skimage.color as color
import numpy as np
import scipy.ndimage as nd
```

```python
image = skimage.data.cat()
```

```python
io.imshow(image)# cat 
io.show()
```

```python
graylion=color.rgb2gray(image)
io.imshow(graylion) #gray cat 
io.show()
```

```python
shiftedlionVert = abs(graylion - nd.shift(graylion,(0,1),order=0)) #subracts shifted-graylion with graylion 
shiftedlionVHori = abs(graylion - nd.shift(graylion,(1,0),order=0)) #subracts shifted-graylion with graylion 
```

```python
io.imshow(-shiftedlionVert, cmap='gray')# "-" flips the color map so edges are black on white.
io.show()
io.imshow(-shiftedlionVHori, cmap='gray')
io.show()
```


## How it works
This method is very simple. It works by subtracting an image's intensity values with a copy of itself that is shifted one pixel vertically or horizontally to highlight areas of greater intensity change. The absolute value of the difference is the changes in intensity value of the image which is synonymous with edges.

Original values: [10, 10, 10, 40, 40, 40, 30, 20, 10]

Shifted Row: [0, 10, 10, 10, 40, 40, 40, 30, 20]

Differences: [10, 0, 0, 30, 0, 0, 0, -10, -10]
