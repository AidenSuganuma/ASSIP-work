# Vertical and horizontel edge detection


## How it works
This method is blessed by ococam's razor. It works by taking the image and subtracting intensity values with a copy of itself shifted one pixel vertically or horizontally to highlight areas of greater intensity change. Of course, the higher intensity values correlate to edges in an image, a common assumption which is seen with other edge detectors like sobel.

```python
import cv2
import skimage
import skimage.io as io
import skimage.filters as filters
import skimage.color as color
import numpy as np
import scipy.ndimage as nd
```

```python
image = skimage.data.cat()
```

```python
io.imshow(image)# cat c:
io.show()
```

```python
graylion=color.rgb2gray(image)
io.imshow(graylion) #gray cat :C
io.show()
```

```python
shiftedlionVert = abs(graylion - nd.shift(graylion,(0,1),order=0)) #subracts shifted-graylion with graylion 
shiftedlionVHori = abs(graylion - nd.shift(graylion,(1,0),order=0)) #subracts shifted-graylion with graylion 
```

abs (absolute value) is used as io.imshow requires [0,1]. furthermore, scipy ndi is sued to shift the array.

```python
io.imshow(-shiftedlionVert, cmap='gray')# "-" flips the color map so edges are black on white.
io.show()
io.imshow(-shiftedlionVHori, cmap='gray')
io.show()
```


## How it works
This is subjectively one of the most straightforward image processing methods so there is not much to expand on. In short, the methodology works by taking a grayscale image and seeing the difference of that image with a copy that is shifted one pixel.
