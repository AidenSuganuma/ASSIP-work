```python
import cv2
import skimage
import skimage.io as io
import skimage.filters as filters
image=skimage.data.cat()
io.imshow(image)
io.show()
```

```python
blur_lion=cv2.blur(image,(11,11))
io.imshow(blur_lion)
io.show()
```

## How it Works
 cv2.blur is an average blurring function. It requires an image and set of odd coordinates. 
The function then takes said coordinates and makes a kernel (a smaller matrix within the image); The blurring algorithm will go chunk by chunk on the image and will use the kernel to dictate the size of each chunk. When evaluating each kernel, cv2.blur will take the center value of each kernel (hence the odd value requirement) and use it as the average. then, it will set every value in that kernel to said average. 



```python
#Original kernel:
#[[0 0 0]
# [0 1 0]
# [0 0 0]]

#cv2.blur kernel:
#[[1 1 1]
 #[1 1 1]
# [1 1 1]]
```

```python
gaussianblur_lion=cv2.GaussianBlur(image,(11,11),0)
io.imshow(gaussianblur_lion)
io.show()
skgaussianblur_lion=filters.gaussian(image)
io.imshow(skgaussianblur_lion)
io.show()
```

## How it works
Gaussian blurring finds the weighted average instead of setting the center pixel to the simple average. This means the average value is calculated based upon how much surrounding pixels contribute to the average.
