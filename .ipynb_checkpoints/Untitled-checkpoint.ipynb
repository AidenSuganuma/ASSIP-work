```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
import skimage
import skimage.io as io
```

```python

# example image
image = np.zeros((100, 100))

# squares
image[10:40, 30:70] = 255
image[55:60, 15:30] = 200
image[70:90, 70:90] = 145


# Add random noise
noise = np.random.normal(0, 15, size=(100, 100))
image += noise

plt.imshow(image, cmap = 'gray')
plt.show
```

```python
x = 2  # Set the power value

# Raise each element of the image array to the power of x
image_powered = np.power(image, x)

```

```python
plt.imshow(image, cmap = 'gray')
plt.show
```

```python
# Demo image 
with rasterio.open('demo.jp2') as src:  #jp2 image 
   
    red = src.read(1, masked=True)  
    green = src.read(2, masked=True)
    blue = src.read(3, masked=True)
image = np.stack([red, green, blue], axis=-1) #Restacking channel 

io.imshow(image)
io.show()
```

```python
gray = skimage.color.rgb2gray(image)
io.imshow(gray)
io.show()
```

```python
demo = gray + 1
```

```python
p = 7
example = np.power(demo, p)
```

```python
norm = example.astype(float) # Converts Bool to float 
norm /= norm.max()
```

```python
io.imshow(norm)
io.show()
```
