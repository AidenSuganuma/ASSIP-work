```python
import rasterio
from rasterio.windows import Window

# Open the original jp2 file
with rasterio.open('T18SUJ_20230420T154819_TCI_10m.jp2') as src:
    # Define the window coordinates for the region of interest
    window = Window(col_off=2600, row_off=2250, width=250, height=250)
    
    # Read the subset of the image corresponding to the window
    subset = src.read(window=window)
    
    # Create a new rasterio dataset for the subset
    subset_dataset = rasterio.open(
        'subset.jp2',
        'w',
        driver='JP2OpenJPEG',
        width=window.width,
        height=window.height,
        count=src.count,
        dtype=subset.dtype,
        crs=src.crs,
        transform=src.window_transform(window)
    )
    
    # Write the subset data to the new dataset
    subset_dataset.write(subset)
    subset_dataset.close()
```

```python
import rasterio
from rasterio.windows import Window

# Open the original JP2 file
with rasterio.open('T18SUJ_20230420T154819_TCI_10m.jp2') as src:
    # Define the window coordinates for the region of interest
    window = Window(col_off=1900, row_off=9100, width=300, height=300)

    # Read the subset of the image corresponding to the window
    subset = src.read(window=window)

    # Create a new rasterio dataset for the subset
    subset_dataset = rasterio.open(
        'demo.jp2',
        'w',
        driver='JP2OpenJPEG',
        width=window.width,
        height=window.height,
        count=src.count,
        dtype=subset.dtype,
        crs=src.crs,
        transform=src.window_transform(window)
    )

    # Write the subset data to the new dataset
    subset_dataset.write(subset)
    subset_dataset.close()
```
