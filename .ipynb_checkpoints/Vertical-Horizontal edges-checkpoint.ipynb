```python
import cv2
import skimage
import skimage.io as io
import skimage.filters as filters
import skimage.color as color
import numpy as np
import scipy.ndimage as nd
import rasterio 
```

```python
# Demo image 
with rasterio.open('demo.jp2') as src:  # jp2 image 
    red = src.read(1, masked=True)  
    green = src.read(2, masked=True)
    blue = src.read(3, masked=True)
image = np.stack([red, green, blue], axis=-1) # Restacking channel 

```

```python
io.imshow(image)
io.show()
```

```python
grayimage = color.rgb2gray(image)
io.imshow(grayimage) # Gray image
io.show()
```

```python
# Subtracts grayimage with a shifted copy
shiftedvert = abs(grayimage - nd.shift(grayimage, (0, 1), order=0))  # Y axis shift
shiftedhori = abs(grayimage - nd.shift(grayimage, (1, 0), order=0))  # X axis shift
```

```python
io.imshow(-shiftedvert, cmap='gray')  # "-" flips the color map so edges are black on white.
io.show()
io.imshow(-shiftedhori, cmap='gray')
io.show()
```


## How it works
This method is very simple. It works by subtracting an image's intensity values with a copy of itself that is shifted one pixel vertically or horizontally to highlight areas of greater intensity change. The absolute value of the difference is the changes in intensity value of the image which is synonymous with edges.

Original values: [10, 10, 10, 40, 40, 40, 30, 20, 10]

Shifted Row: [0, 10, 10, 10, 40, 40, 40, 30, 20]

Differences: [10, 0, 0, 30, 0, 0, 0, -10, -10]
