```python
import cv2
import skimage
import skimage.io as io
import numpy as np
from skimage import transform
import skimage.filters as filters
```

```python
def flipblur(image, flip_code):
    if flip_code == 0:  # Flip horizontally
        temp = np.fliplr(image)
    elif flip_code == 1:  # Flip vertically
        temp = np.flipud(image)
    elif flip_code == -1:  # Flip both vertically and horizontally
        temp = np.flipud(np.fliplr(image))
    else:
        raise ValueError("use 0, -1, 1")
    return filters.gaussian(temp)

```

```python
lion = skimage.data.cat()
flipblurlion = flipblur(lion, -1)#change intiger to different flip code if needed 
io.imshow(flipblurlion) 
io.show()
```

## How it Works
It is simply the flip function we made but with a guassian at the end.
