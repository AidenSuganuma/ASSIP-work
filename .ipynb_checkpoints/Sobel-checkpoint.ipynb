# Sobel 
The Sobel operator is commonly used for edge detection. Named after Irwin Sobel who introduced the operator in 1968, the operator is designed to highlight regions of an image where the intensity values are the strongest, which often corresponds to the presence of edges. This operator's base functionality works on a kernel system (like blur) where it analyzes bits of the image in 3x3 chunks. It can be run on either axis or both at the same time; When run on the latter, a color gradient is often used to highlight the direction of which the edge is facing. 


```python
import cv2
import skimage
import skimage.io as io
import skimage.filters as filters
import skimage.color as color
image=skimage.data.cat()
io.imshow(image)
io.show()
```

```python
#try to run this code.
#lionxy = filters.sobel_v(image)
```


## How it Works
Sobel x and y (H(orizontal) and V(ertical) for skimage) works with intensity values of a single channel, because the image is RGB, it has 3 intensity channels and will not be able to evaluate it. we must use grayscale images for sobel X and Y to reduce to one channel.


```python
grayimage = color.rgb2gray(image)
io.imshow(grayimage)
io.show()
```

```python
lionx = filters.sobel_h(grayimage)
liony = filters.sobel_v(grayimage)
io.imshow(lionx)
io.show
```

```python
io.imshow(liony)
io.show
```

## How it Works
Now you can see the difference between sobel X and Y. one highlights the vertical hairs on the cat's head while ignoring the horizontal whiskers while the other does the opposite; this is a result of the differing directions of analysis done by sobel.
## SobelXY
As expected, sobel XY will combine for the best of both worlds. Rather unexpectedly, however, the skimage's sobel function (filters.sobel) combines X Y AND takes into account the possibility for multiple channels, allowing for it to take in RGB images directly. (it does sobel to each channel and restacks) 


```python
graylionxy = filters.sobel(grayimage)
io.imshow(graylionxy)
io.show()
```

```python
lionxy = filters.sobel(image)
io.imshow(lionxy)
io.show()
```
